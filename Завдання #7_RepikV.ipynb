1. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{(y_{i} - (b + mx_{i}))^{2}}$$

&emsp;&emsp;де x,y - деякі задані вектори

```python
import numpy as np

np.random.seed(42)
x = np.random.normal(0, 5, 100)
y = 50 + 2 * x + np.random.normal(0, 2, len(x))
```

```python
def fun1(b, m):
    return np.mean((y - (b + m*x))**2)

def df(b, m):
    h = 1e-10
    df_b = (fun1(b + h, m) - fun1(b, m)) / h
    df_m = (fun1(b, m + h) - fun1(b, m)) / h
    return np.array([df_b, df_m])
```

```python
wb = np.ones(2)

alpha = 0.01          # крок алгоритма
epsilon = 0.00001     # якщо зміна w менше "epsilon", то що рішення знайдено

while True:
    x1 = wb[0]
    y1 = wb[1]
    w = wb - alpha * df(x1, y1)  # крок алгоритму
    
    if np.sum(np.abs(w - wb)) <= epsilon:  # перевірка (і можливо пошук припинити)
        break
    else:
        wb = w  # оновлення попереднього значення
            
    print(w, fun1(w[0], w[1]))  # виведення поточного значення
    

print('=' * 100)
print('Значення b і m, при якому f набуває мінімального значення: ', w)
print('Мінімум f: ', fun1(w[0], w[1]))

  
```

2. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{[(y_{i} - (b + mx_{i}))^{2}]} + m^2 + b^2$$

&emsp;&emsp;де x,y - вектори з 1-го завдання

```python
def fun2(b, m):
    n = len(x)
    return np.sum((y - (b + m*x))**2) / n + m**2 + b**2

def df2(b, m):
    h = 1e-10
    f_x = (fun2(b + h, m) - fun2(b, m)) / h
    f_y = (fun2(b, m + h) - fun2(b, m)) / h
    return np.array([f_x, f_y])

wb = np.ones(2)
alpha = 0.001          # крок алгоритма
epsilon = 0.00001     # критерій, якщо зміна w менше "epsilon", то вважаємо, що рішення знайдено

while True:
    x1 = wb[0]
    y1 = wb[1]
    w = wb - alpha * df2(x1, y1)  # крок алгоритму, перехід до наступного значення w
    
    if np.sum(np.abs(w - wb)) <= epsilon:  # перевірка (тут, можливо, пошук варто припинити)
        break
    else:
        wb = w  # оновлення попереднього значення
            
    print(w, fun2(w[0], w[1]))  # виведення поточного значення х та y
    

print('=' * 100)
print('Значення b і m, при якому f набуває мінімального значення: ', w)
print('Мінімум f: ', fun1(w[0], w[1]))
```
