```python
# Завантаження модулів
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\Vlad\\Desktop\\PYTHON\\Заняття №13\diabetes.csv', sep=',', header=0)
# data.info()

# Назви колонок
# Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age,Outcome

# SVM
# вибираємо корисні дані
data = data[(data['BloodPressure'] > 0) & (data['Glucose'] > 0) & (data['BMI'] > 0) & (data['Insulin'] > 0)]

data.info()
X = data[['Age', 'Glucose', 'BMI']].values
y = data['Outcome'].values  
# data.head()
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.show()
# по графіку бачимо що з віком рівень глюкози в крові підвищується
```

```python
plt.scatter(X[:, 1], X[:,2], c=y)
plt.xlabel('Glucose')
plt.ylabel('BM index')
plt.show()
# по графіку бачимо, худіші люди мають менший рівень глюкози
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, # перший масив, який потрібно розбити
    y, # другий масив, який потрібно розбити
    test_size=0.2, # розмір меншої групи (20% від загальної кількості даних)
    random_state=12 # фіксація випадкового стану: тепер при перезапуску групи будуть однаковими
)

print('Розмір вибірки для навчання:', len(X_train))
print('Розмір вибірки для тестування:', len(X_test))
```

```python
from sklearn.linear_model import LinearRegression

# ініціалізація моделі
mdl = LinearRegression(
    fit_intercept=True # чи включити в модель константу "с" з формули вище
)

# навчання моделі
mdl = mdl.fit(X_train, y_train)
```

```python
print('constant term: ', mdl.intercept_)
```

```python
from sklearn.metrics import mean_squared_error

mean_forecast = np.ones_like(y_test) * np.mean(y_train) 
model_forecast = mdl.predict(X_test) 

print('Помилка прогнозу, при середньому значенні:', mean_squared_error(y_test, mean_forecast))
print('Помилка прогнозу з використанням моделі: ', mean_squared_error(y_test, model_forecast))
```

```python
y_train_forecast = mdl.predict(X_train)

print('Помилка на навчальних даних: ', mean_squared_error(y_train, y_train_forecast))
print('Помилка тестових даних: ', mean_squared_error(y_test, model_forecast))
```

```python
# Логістична регресія

from sklearn.linear_model import LogisticRegression
mdl = LogisticRegression(fit_intercept=True).fit(X_train, y_train)
```

```python
  # Колонки 'Age', 'Glucose', 'BMI'
print('Glucose: ', mdl.coef_[0][1])
print('BMI: ', mdl.coef_[0][2])
```

```python
from sklearn.metrics import accuracy_score

predictions = mdl.predict(X_test) # отримаємо прогнози з моделі
print('Оцінна точність: ', accuracy_score(y_test, predictions))
```

```python
# Пошук кращого параметра за допомогою крос-валідації

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

svm = SVC()  # ініціалізація моделі

# ініціалізація значень параметрів, для яких здійснити пошук (будуть протестовані всі комбінації значень)
params_grid = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [0.5, 1, 2, 5]
}

# ініціалізація об'єкта для пошуку оптимальних значень параметрів
clf = GridSearchCV(
    estimator=svm,  # модель, яку потрібно протестувати
    param_grid=params_grid,  # словник з іменами параметрів та списком їх значень
    scoring='f1_macro',  # назва метрики якості, яку необхідно використовувати
    cv=4  # кількість тестових наборів
)

clf = clf.fit(X, y)  # пошук значень
grid_search_results = pd.DataFrame(clf.cv_results_)  # отримання результатів пошуку у вигляді таблиці
grid_search_results
```

Виберем найкращий параметр по Score=1 (C': 2, 'kernel': 'poly)

Метод опорних векторів

```python
from mlxtend.plotting import plot_decision_regions

X1 = data[['Age', 'Glucose']].values   
## X1 = data[['BMI', 'Glucose']].values

# ініціалізувати модель SVM (в sklearn SVC - support vectors classifier)
svm = SVC(
    C=2.,  # регулярізація 
    kernel='poly'  # тип відстані між точками (про цей параметр нижче) linear
)

svm = svm.fit(X1, y)  # навчимо модель

# побудуємо графік, де відобразимо в яких областях який клас вибирає наша модель
plot_decision_regions(X1, y, clf=svm)
plt.show()
```

```python
from sklearn.metrics import f1_score

y_preds = svm.predict(X1)
f1_score(y, y_preds)
```

```python
# NBC
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB().fit(X1, y)  # визначення та тренування класифікатора

# побудуємо графік із областями прийняття рішень
plot_decision_regions(X1, y, clf=gnb)
plt.show()
```

```python
y_preds = gnb.predict(X1)
f1_score(y, y_preds, average='macro')
```
